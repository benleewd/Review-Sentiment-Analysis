```python
def save_deal_csv(df_input,file_address):
    df_unique=df_input.copy(deep=True)
    try:
        print("Saving...")
       # print(df_unique.describe)
        df_unique.to_csv(file_address, index=False,encoding='utf-8-sig')
        print("Done.")
        print('Saved as '+file_address)    
    except (RuntimeError, TypeError, NameError ,PermissionError) as e:
        print(e)
```

```python
def load_downloaded_files(path_input):
    path = path_input
    # combine files 
    print(path)
    all_files = glob.glob(path + "*.csv")
    file_counter=0
    li = []

    for filename in all_files:
        file_counter+=1
        print(filename+" added ")
        df = pd.read_csv(filename, index_col=None, header=0, delimiter = ',',encoding='utf-8-sig')
        print("with "+ str(len(df))+" rows \n")
        li.append(df)


    df = pd.concat(li, axis=0, ignore_index=True,sort=False)
    print("All files has been loaded -  total : "+str(file_counter)+" files   Done!")
    print("with "+ str(len(df))+" rows \n")
    return df

```

```python
def clean_str(x):
    
    result = re.sub(r'[^a-zA-Z]', " ", x) 
    x_list=result.split()
    x_string=""
    if len(x_list)>=0:
        
        for x_word in x_list:
            if(len(x_word))>2 and (len(x_word))<15:
                x_string+=(x_word+" ")
    x_list=[]
        
    return (x_string.strip().lower())
```

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

def get_lemmatization (x):
    sentence = x
    x_string=''
    sentence_words = word_tokenize(sentence)
    for word in sentence_words:
        x_string+=(wordnet_lemmatizer.lemmatize(word)+" ")
    return x_string.strip()
```

Key hyperparameters we can choose are

The model type (continuous bag of words - surrounding words predicting to word they flank. Skipgram - flanked word predicting all its surrounding words) skipgram typically is favored for large corpus
Sampling method ( hierarchical or negative ) again latter typically preferred for large corpus
Iterations - 5 or even as low as 3 for large corpus of 30 million unique words or more
Dimensions - 300 works for corpus of upto 60 million unique words
Subsampling - determines how often frequently occurring words like “the” are trained. Could be anywhere from 1e-3 to 1e-5. Values lower than 1e-5 tends to influence vector quality
Window size - smaller window size gives results that are more syntactic in nature. Larger window ( size > 5 ) gives results that are more semantic - this takes more training time given larger window size.

```python
max_features = 500000
embedding_dim=300
def get_emb(num):
    embeddings_index = {}
    embeddings_index2= {}
    if num==1:
        f = open(os.path.join('./vec_input/', en_vec))
        for line in f:
            values = line.split()
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_index[word] = coefs
        f.close()

        print('Found %s after adding ENGLISH word vectors.' % len(embeddings_index))

        f = open(os.path.join('./vec_input/', custom_vec))
   
        for line in f:
            #values = line.split("\r\n")
            values = line.split()
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_index[word] = coefs
        f.close()

        print('Found %s after adding self generated word vectors.' % len(embeddings_index))
         
        
        word_index = tokenizer.word_index
        print('Found %s unique tokens.' % len(word_index))

        num_words = min(max_features, len(word_index)) + 1   
    
        embedding_matrix = np.zeros((len(word_index) + 1, embedding_dim))

        # for each word in out tokenizer lets try to find that work in our w2v model
        for word, i in word_index.items():
            if i > max_features:
                continue
            embedding_vector = embeddings_index.get(word)
            if embedding_vector is not None:
                # we found the word - add that words vector to the matrix
                embedding_matrix[i] = embedding_vector
            else:
                # doesn't exist, assign a random vector
                embedding_matrix[i] = np.random.randn(embedding_dim)
        return (embedding_matrix,embeddings_index,num_words)
    
    
    
    

    elif num==2:
        
        f = open(os.path.join('./vec_input/', en_vec))
     
        for line in f:
            values = line.split()
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_index2[word] = coefs
        f.close()

        print('Found %s after adding ENGLISH word vectors.' % len(embeddings_index2))

        f = open(os.path.join('./vec_input/', indo_vec))
        for line in f:
            values = line.split("\r\n")
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_index2[word] = coefs
        f.close()

        print('Found %s after adding INDO word vectors.' % len(embeddings_index2))
        
        f = open(os.path.join('./vec_input/', custom_vec))

        for line in f:
            values = line.split()
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_index2[word] = coefs
        f.close()

        print('Found %s after adding custom_glove word vectors.' % len(embeddings_index2))


        word_index2 = tokenizer.word_index
        print('Found %s unique tokens.' % len(word_index2))

        num_words = min(max_features, len(word_index2)) + 1

        
        
        embedding_matrix2 = np.zeros((len(word_index2) + 1, embedding_dim))

        # for each word in out tokenizer lets try to find that work in our w2v model
        for word, i in word_index2.items():
            if i > max_features:
                continue
            embedding_vector = embeddings_index2.get(word)
            if embedding_vector is not None:
                # we found the word - add that words vector to the matrix
                embedding_matrix2[i] = embedding_vector
            else:
                # doesn't exist, assign a random vector
                embedding_matrix2[i] = np.random.randn(embedding_dim)
        return (embedding_matrix2,embeddings_index2,num_words,word)
```

```python
def average_word_vectors(words, model, vocabulary, num_features):
    
    feature_vector = np.zeros((num_features,),dtype="float64")
    nwords = 0.
    
    for word in words:
        if word in vocabulary: 
            nwords = nwords + 1.
            feature_vector = np.add(feature_vector, model[word])
    
    if nwords:
        feature_vector = np.divide(feature_vector, nwords)
        
    return feature_vector


def averaged_word_vectorizer(corpus, model, num_features):
    vocabulary = set(model.wv.index2word)
    features = [average_word_vectors(tokenized_sentence, model, vocabulary, num_features)
                    for tokenized_sentence in corpus]
    return np.array(features)

```
